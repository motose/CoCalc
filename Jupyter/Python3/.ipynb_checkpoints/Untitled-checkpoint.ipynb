```python
import matplotlib.pyplot as plt

price = [100, 250, 380, 500, 700]
number = [1, 2, 3, 4, 5]

# 
plt.plot(price, number)

# 
plt.title("price / number")

# x
plt.xlabel("price")

# y
plt.ylabel("number")

# 
plt.show()
```

```python
def fib(i):
    x,y = 0,1
    for i in range(0,i):
        x,y = y,x+y
    return x
```

```python
fib(39)
```
